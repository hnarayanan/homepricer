```python
import math
import random

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

%matplotlib inline
```

```python
N = 1500
noise = 0.0
```

```python
positions = (2.0*np.random.rand(N, 2) - 1.0)
x = positions[:, 0]
y = positions[:, 1]
```

```python
def f(x, y):
    r = math.sin(10*(x**2 + y**2))
    if r <= 0.5 + noise*(random.random() - 0.5)*2.0:
        return 0.0
    else:
        return 1.0
```

```python
r = np.empty(N)
for i in range(N):
    r[i] = f(x[i], y[i])
```

```python
plt.axis([-1, 1, -1, 1])
plt.scatter(x, y, c=r, alpha=0.75, s=50, edgecolors='none')
```

```python
X = positions
y = r
```

```python
clf = MLPClassifier(activation='tanh', hidden_layer_sizes=(2000, 2000), random_state=1)
```

```python
clf.fit(X, y)
```

```python
test_positions = (2.0*np.random.rand(10000, 2) - 1.0)
test_predictions = clf.predict(test_positions)
```

```python
plt.axis([-1, 1, -1, 1])
plt.scatter(test_positions[:, 0], test_positions[:, 1], c=test_predictions, alpha=1.0, s=4, edgecolors='none')
```
